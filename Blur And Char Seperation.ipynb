```python
import tesseract
import ctypes
import os
```

```python
api = tesseract.TessBaseAPI()
api.SetOutputName("outputName");
```

```python
api.Init(".","eng",tesseract.OEM_DEFAULT)
api.SetPageSegMode(tesseract.PSM_AUTO)
```

```python
#fileimage = './medimages/simpletextsparse.jpg'  # OR
fileimage = './medimages/simpletextdense.jpg'
```

```python
Image(filename=fileimage)
```

```python
mImgFile = fileimage
```

```python
result = tesseract.ProcessPagesWrapper(mImgFile,api)
print "result(ProcessPagesWrapper)=",result
```

```python
from PIL import Image
from PIL import ImageFilter
from scipy import ndimage
original_image = Image.open(fileimage)
blurred_image = ndimage.filters.gaussian_filter(original_image,1)
j = Image.fromarray(blurred_image)
j.save("./testocrimageblur.png")
```
